```
import random

player1 = str(input("Name of Player 1: "))
player2 = str(input("Name of Player 2: "))
player3 = str(input("Name of Player 3: "))
player4 = str(input("Name of Player 4: "))
print("\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n\n")
category = ["Raja", "Mantri", "Sipai", "Chor"]
result1 = 0
result2 = 0
result3 = 0
result4 = 0
for i in range(1, 6):
    player1_draw = random.randint(0, 3)
    player2_draw = random.randint(0, 3)
    player3_draw = random.randint(0, 3)
    player4_draw = random.randint(0, 3)
    while player1_draw == player2_draw or player1_draw == player3_draw or player1_draw == player4_draw or player2_draw == player3_draw or player2_draw == player4_draw or player3_draw == player4_draw:
        player1_draw = random.randint(0, 3)
        player2_draw = random.randint(0, 3)
        player3_draw = random.randint(0, 3)
        player4_draw = random.randint(0, 3)
    if player1_draw == 0:
      result1 = result1 + 1000
    elif player1_draw == 1:
      result1 = result1 + 500
    elif player1_draw == 2:
      result1 = result1 + 100 
    if player2_draw == 0:
      result2 = result2 + 1000
    elif player2_draw == 1:
      result2 = result2 + 500
    elif player2_draw == 2:
      result2 = result2 + 100  
    if player3_draw == 0:
      result3 = result3 + 1000
    elif player3_draw == 1:
      result3 = result3 + 500
    elif player3_draw == 2:
      result3 = result3 + 100  
    if player4_draw == 0:
      result4 = result4 + 1000
    elif player4_draw == 1:
      result4 = result4 + 500
    elif player4_draw == 2:
      result4 = result4 + 100  
    print("Result of Draw "+ str(i) + " " + category[player1_draw] + ',' + category[player2_draw] + ',' + category[player3_draw] + ',' + category[player4_draw])        

print("\n\nScores \n" + player1 + ":" + str(result1) + " " + player2 + ":" + str(result2) + " " + player3 + ":" + str(result3) + " " + player4 + ":" + str(result4))
if result1 > result2 and result1 > result3 and result1 > result4:
  print("\n" + player1 + " is the winner!")
elif result2 > result1 and result2 > result3 and result2 > result4:
  print("\n" + player2 + " is the winner!")
elif result3 > result1 and result3 > result2 and result3 > result4:
  print("\n" + player3 + " is the winner!")
elif result4 > result1 and result4 > result3 and result4 > result2:
  print("\n" + player4 + " is the winner!")

if result1 == result2 or result1 == result3 or result1 == result4 or result2 == result3 or result2 == result4 or result3 == result4: 
  print("It's a tie!")
```
